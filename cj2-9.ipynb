```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pull 1000 random numbers distributed uniformly between [0,1]
x = np.random.uniform(0.0,1.0,1000)
#Histogram the random numbers into 100 bins, and plot the histogram.

plt.hist(x, edgecolor="black", bins=100) 
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
plt.savefig('plot.pdf')
```
