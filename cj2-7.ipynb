```python
x = np.linspace(0,1,100)

def exp_function(x):
    return np.exp(x)

#function to set y=exp(x), and then plot x vs. y
def set_plot():
    y = exp_function(x)
    #Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. 
    plt.xlabel('Time[millseconds]')
    plt.ylabel('Awesomeness')
    #Save the figure as a PDF.
    plt.savefig('plot.pdf')
    plt.plot(x,y)
    plt.show()

set_plot()
```
