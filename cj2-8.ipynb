```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2)

x = np.linspace(0, 1, 100)  # create array from [0,1] in 100 steps
y = np.sin(x)
w = np.cos(x)

axes[0].scatter(x, y)
axes[1].scatter(x, w)

axes[0].set_xlabel('X')
axes[0].set_ylabel('sin(X)')

axes[1].set_xlabel('X')
axes[1].set_ylabel('cos(X)')

    #Save the figure as a PDF.
plt.savefig('plot.pdf')

plt.show()
```
