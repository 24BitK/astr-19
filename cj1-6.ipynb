```python
import numpy as np

def run_comp():
    num_entries = 1000
    start_x = 0
    stop_x = 2 * np.pi
    hit = (stop_x - start_x) / (num_entries - 1)

    # print table
    print("{:15} {:15}".format('x', 'sin(x)'))
    print("-" * 35)

    #iterate
    for i in np.arange(start_x, stop_x, hit):

        x = i
        cos_value = np.cos(x)
        sin_value = np.sin(x)

        print("{:15} {:15} ".format(x,sin_value))
        print("{:15} {:15} ".format(x,cos_value))

if __name__ == '__main__':
    run_comp()
```
